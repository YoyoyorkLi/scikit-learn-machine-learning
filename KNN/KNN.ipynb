```python
import pandas as pd
import pandas as pd
from sklearn import preprocessing, neighbors, metrics
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('KNN_dataset.csv')
df
```

```python
train_df = df.iloc[0:100]
test_df = df.iloc[100:175]
test_df
```

```python
columns = ['col0','col1','col2','col3','col4','col5','col6','col7','col8','col9','col10', 'col11', 'col12']
train_X = train_df[columns]
train_y = train_df['label']
test_X = test_df[columns]
test_y = test_df['label']

train_X
```

```python
scaler = StandardScaler()
scaler.fit(train_X)
train_X = scaler.transform(train_X)
test_X = scaler.transform(test_X)
train_X
```

```python
model = neighbors.KNeighborsClassifier(5)
model.fit(train_X, train_y)
predictions = model.predict(test_X)

(predictions == test_y).mean()
predictions
```

```python
acc = metrics.cohen_kappa_score(test_df['label'], predictions)
acc
```
